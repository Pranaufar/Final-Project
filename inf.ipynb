```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('application_test.csv')
```

```python
df
```

```python
# Load the model pipeline
loaded_model = joblib.load('model_pipeline.pkl')

# Use the loaded model pipeline for predictions
y_pred = loaded_model.predict(df)
```

```python
y_pred.shape
```

```python
y_pred.sum()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay , classification_report

report = classification_report(df, y_pred)
print(report)

cm = confusion_matrix(df, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

plt.show()
```
